```python
# Importing libraries
import csv
import numpy as np
from sklearn.metrics import cohen_kappa_score
```

```python
# Ground truth definition
n = 23
gt = [1] * n
gt[3] = 0
```

```python
# Analyzing Cohen Kappa for different thresholds
def analyze(alias, th):
  values = [0] * n

  with open("/home/logprobs.csv", mode="r", newline="\n", encoding="utf-8") as handle:
      reader = csv.DictReader(handle)
      for row in reader:
        category = row["category"]
        principle = int(row["principle"])
        goal = int(row["goal"])
        logprob = float(row["logprob"])

        if category == alias and np.exp(logprob) >= th:
          values[goal] = 1

  return cohen_kappa_score(gt, values)
```

```python
# Printing results
thss = [x / 100 for x in range(0, 101, 5)]
thresholds = ""
kappas = ""

for ths in thss:
  kappa = analyze("GDPR (Europe)", ths)
  thresholds = thresholds + str(ths)
  kappas = kappas + str(kappa)
  if ths < 1:
    thresholds = thresholds + ","
    kappas = kappas + ","

print(thresholds)
print(kappas)
```
